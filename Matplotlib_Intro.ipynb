### Comparing the pyplot method with object-oriented plotting

```python
# The first method, the original one that creates an empty figure (pyplot)
fig = plt.figure() # creates an empty figure
ax = fig.add_subplot() # adds some axes
plt.show();
```

```python
# The second method that creates some data
fig = plt.figure() # creates an empty figure
ax = fig.add_axes([1, 1, 1, 1])
x = [1, 2, 3, 4]
y = [22, 33, 44, 55]
ax.plot(x, y) # add in the data
plt.show()
```

```python
# The third method, object-oriented method, the one that will be used most of the time
fig, ax = plt.subplots()
x = [2, 4, 6, 8]
y = [50, 100, 110, 200]
ax.plot(x, y)
```
